```python
data.drop(['Unnamed: 0'], axis=1,inplace=True)
```

```python
Starting_years = data['Starting Year'].value_counts()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=Starting_years.index, y=Starting_years.values)
plt.title("No. of startups according to year", fontdict={"fontsize":15})
plt.xlabel("Years",fontdict={"fontsize":15,"color":"b"})
plt.ylabel("No. of startups",fontdict={"fontsize":15,"color":"b"})
plt.show()
```

```python
# top 10 companies with highest funding rounds
plt.figure(figsize=(8,6))
sns.barplot(x='Funding Round',y='Company',data=data.sort_values('Funding Round', ascending=False).iloc[:10],alpha=0.8)  
plt.title('Companies with highest funding rounds',fontsize=20)    #Note labelling the data
plt.ylabel('Company',fontsize=20)               #Note labelling the y-label
plt.xlabel('Count',fontsize=20)               #Note labelling the x-label
plt.tick_params(labelsize=8)
plt.show()
```

```python
a1=data["No. of Investors"].sort_values(ascending=False).head(10)
a1
```

```python
a2=data.loc[[11, 23, 50, 10, 32, 69, 31, 88, 4, 13],"Company"]
a2
```

```python
sns.barplot(x=a2,y=a1,data=data)
plt.xticks(rotation='vertical')
plt.title('Top 10 Companies with highest no. of Investors',size=20)
plt.xlabel('Company',size=15)
plt.ylabel('No. of Investors',size=15)
plt.tick_params(labelsize='x-small')
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Funding Amount in $',y='Company',data=data.sort_values('Funding Amount in $', ascending=False).iloc[:10],alpha=0.8)  
plt.title('Companies with highest funding ',fontsize=20)    #Note labelling the data
plt.ylabel('Company',fontsize=20)               #Note labelling the y-label
plt.xlabel('Funding Amount in $',fontsize=20)               #Note labelling the x-label
plt.tick_params(labelsize=8)
plt.show()
```

```python
#highest startup in  city

plt.figure(figsize=(10,7))
data.groupby('City')['Company'].count().sort_values(ascending=False).plot.bar();
plt.title("Cities with highest no. of startups")
plt.xticks(rotation='vertical')
plt.ylabel("No. of Startups")

plt.show()
```
